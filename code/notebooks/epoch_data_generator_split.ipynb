```python
import os
import sys
import pandas as pd
from sklearn.model_selection import train_test_split

sys.path.append("..")
from utils.epoch_data_generator import EpochDataGenerator
```

```python
participants_file = os.path.join(EpochDataGenerator.data_dir(), "participants.tsv")
participants_df = pd.read_csv(participants_file, sep="\t")
subjects = participants_df.loc[participants_df["exclude"] == 0, "participant_id"].to_list()

subject_epoch_coordinates = [[subject, coordinate] for subject in subjects for coordinate in range(24648)]
train_coordinates, validation_coordinates = train_test_split(subject_epoch_coordinates, test_size=0.2, random_state=42)

train_generator = EpochDataGenerator(train_coordinates)
validation_generator = EpochDataGenerator(validation_coordinates)
```
